```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a, c, b = 0, 0.9, 3
def f(x: np.float64) -> np.float64:
    return 6*np.exp(-x) * np.sin(2*np.pi*x)

def df(x: np.float64) -> np.float64:
    return 12*np.pi*np.exp(-x)*np.cos(2*np.pi*x) - 6*np.exp(-x)*np.sin(2*np.pi*x)

def d2f(x: np.float64) -> np.float64:
    return (6*np.exp(-x)*np.sin(2*np.pi*x) -
            24*np.pi*np.exp(-x)*np.cos(2*np.pi*x) - 
            24 * np.pi**2 *np.exp(-x)*np.sin(2*np.pi*x))
```

```python
def leftDiff(x: np.float64, h: np.float64) -> np.float64:
    return (f(x) - f(x - h))/h

def rightDiff(x: np.float64, h: np.float64) -> np.float64:
    return (f(x + h) - f(x))/h

def d2fApprox(x: np.float64, h: np.float64) -> np.float64:
    return (f(x + h) - 2*f(x) + f(x - h))/h**2
```

```python
x_data = range(1, 15)
fig, axs = plt.subplots(1, 2, figsize=(15, 5))
axs[0].plot(x_data, [df(c) for i in x_data], label="$f'(x)$", color="red")
axs[0].plot(x_data, [leftDiff(c, np.power(10.0, -i)) for i in x_data], label="$leftDiff$", color="blue")
axs[0].plot(x_data, [rightDiff(c, np.power(10.0, -i)) for i in x_data], label="$rightDiff$", color="green")

axs[1].plot(x_data, [np.abs(df(c) - leftDiff(c, np.power(10.0, -i))) for i in x_data], label="$leftDiff$", color="blue")
axs[1].plot(x_data, [np.abs(df(c) - rightDiff(c, np.power(10.0, -i))) for i in x_data], label="$rightDiff$", color="green")
ax = plt.gca()
axs[0].legend()
axs[0].grid(True)
axs[1].grid(True)
ax = plt.gca()
ax.axhline(y=0, color='k')
plt.yscale('log')
axs[1].legend()
```

```python
x_data = range(1, 15)
fig, axs = plt.subplots(1, 2, figsize=(15, 5))
axs[0].plot(x_data, [d2f(c) for i in x_data], label="$f''(x)$", color="red")
axs[0].plot(x_data, [d2fApprox(c, np.power(10.0, -i)) for i in x_data], color="blue")

axs[1].plot(x_data, [np.abs(d2f(c) - d2fApprox(c, np.power(10.0, -i))) for i in x_data], color="green")
ax = plt.gca()
axs[0].legend()
axs[0].grid(True)
axs[1].grid(True)
ax = plt.gca()
ax.axhline(y=0, color='k')
plt.yscale('log')
```
